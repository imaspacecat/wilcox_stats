```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import re

uniTxt = ["unis2019.txt", "unis2020.txt", "unis2021.txt", "unis2022.txt"] 

unisS = ""
for uniName in uniTxt:
    with open(uniName) as f:
        unisS += f.read()
        
unisS = re.sub('\n', '', unisS)
unisList = re.findall('[^\(\)]+', unisS)

# unisDict = {}
unisDict = dict.fromkeys([unisList[i] for i in range(0, len(unisList), 2)], 0)

for i in range(0, len(unisList), 2):
    unisDict[unisList[i]] += int(unisList[i+1])

# print(unisDict)

unisDict = {k: v for k, v in sorted(unisDict.items(), key=lambda item: item[1])}
# print(uniDic)
names = list(unisDict.keys())
values = list(unisDict.values())

plt.figure(figsize=(70,15))
plt.bar(range(len(unisDict)), values, tick_label=names)
plt.xticks(rotation=90)
plt.show()


```
